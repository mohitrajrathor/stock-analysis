```python
from tvDatafeed import TvDatafeed, Interval 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading SBIN data from 14-Jun-2022 to 14-Jun-2023.
tv = TvDatafeed('mohitrathor7729', 'Mohit7729@#')
data = tv.get_hist(symbol= "SBIN", exchange="NSE", interval= Interval.in_daily, n_bars= 1000)
data
```

```python
# making volatility columns.
data['Volatility'] = (data['high'] - data['low']) / data['low'] * 100
data['Volatility'] = data['Volatility'].round(2)
data
```

```python
# getting rows where Volatility higher than 2%
data[(data['Volatility'] > 2) | (data['Volatility'] < -2)]
```

```python
# getting rows where volume higher tha n 3 months avg.
data[data['volume'] > data['volume'].rolling(60).mean()]
```
