```python
# calculating moving averages
def SMA(col: pd.Series, period: int = 50):
    return col.rolling(period).mean()
```

```python
# calculate moving average
data["MA20"] = SMA(data['Close'], 20)
data['MA50'] = SMA(data['Close'])
data["MA200"] = SMA(data['Close'], 200)
data
```

```python
def RSI(change : pd.Series, period: int):
    gain = change.where(change > 0, 0)
    loss = change.where(change < 0 , 0)
    loss = loss.abs()
    avgain = gain.rolling(period).mean()
    avloss = loss.rolling(period).mean()
    rs = avgain / avloss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
data["RSI"] = RSI(data['Close']-data["Open"], 14)
```

```python
def ADX14(data):
    data['TR'] = pd.DataFrame({
        'a': data['High'] - data['Low'],
        'b': abs(data['High'] - data['Close'].shift()),
        'c': abs(data['Low'] - data['Close'].shift())
    }).max(axis=1)

    # Calculate the Directional Movement (DM)
    data['DM+'] = (data['High'] - data['High'].shift()) > (data['Low'].shift() - data['Low'])
    data['DM-'] = (data['Low'].shift() - data['Low']) > (data['High'] - data['High'].shift())

    # Calculate the True Directional Indicator (+DI) and (-DI)
    TR14 = data['TR'].rolling(window=14).sum()
    DM14plus = data['DM+'].rolling(window=14).sum()
    DM14minus = data['DM-'].rolling(window=14).sum()

    DIplus14 = DM14plus / TR14 * 100
    DIminus14 = DM14minus / TR14 * 100

    # Calculate the Directional Movement Index (DMI)
    DX14 = abs(DIplus14 - DIminus14) / (DIplus14 + DIminus14) * 100
    ADX14 = DX14.rolling(window=14).mean()
    return ADX14
```

```python
def ADX14(high, low, close):
    TR = pd.DataFrame({
        'a': high - low,
        'b': abs(high - close.shift()),
        'c': abs(low - close.shift())
    }).max(axis=1)

    # Calculate the Directional Movement (DM)
    DMplus = (high - high.shift()) > (low.shift() - low)
    DMminus = (low.shift() - low) > (high - high.shift())

    # Calculate the True Directional Indicator (+DI) and (-DI)
    TR14 = TR.rolling(window=14).sum()
    DM14plus = DMplus.rolling(window=14).sum()
    DM14minus = DMminus.rolling(window=14).sum()

    DIplus14 = DM14plus / TR14 * 100
    DIminus14 = DM14minus / TR14 * 100

    # Calculate the Directional Movement Index (DMI)
    DX14 = abs(DIplus14 - DIminus14) / (DIplus14 + DIminus14) * 100
    ADX14 = DX14.rolling(window=14).mean()
    return ADX14
```

```python
data['ADX14'] = ADX14(data['High'], data['Low'], data["Close"]).round(2)
data['RSI'] = RSI(data['Close']-data["Open"], 14).round(2)
data['SMA50'] = SMA(data['Close']).round(2)
```
