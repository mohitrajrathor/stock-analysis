```python
# calculating moving averages
def SMA(col: pd.Series, period: int = 50):
    return col.rolling(period).mean()
```

```python
# calculate moving average
data["MA20"] = SMA(data['Close'], 20)
data['MA50'] = SMA(data['Close'])
data["MA200"] = SMA(data['Close'], 200)
data
```

```python
# making RSI
# def RSI(data: pd.DataFrame, period:int):
change = (data['Close'] - data["Open"])
data["gain"] = change.where(change > 0, 0)
# print(gain)
data["loss"] = change.where(change < 0 , 0)
data["loss"] = data["loss"].abs()
data["av gain"] = data["gain"].rolling(14).mean()
data["av loss"] = data["loss"].rolling(14).mean()

rs = data["av gain"] / data["av loss"]
# print(rs)
data["RSI"] = 100 - (100 / (1 + rs))
data.drop(['gain', 'loss', 'av gain', 'av loss'], axis = 1)
data.tail(50)
```

```python
def RSI(change : pd.Series, period: int):
    gain = change.where(change > 0, 0)
    loss = change.where(change < 0 , 0)
    loss = loss.abs()
    avgain = gain.rolling(period).mean()
    avloss = loss.rolling(period).mean()
    rs = avgain / avloss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
data["RSI"] = RSI(data['Close']-data["Open"], 14)
```
